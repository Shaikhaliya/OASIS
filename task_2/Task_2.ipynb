```python
import pandas as pd
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Unemployment in India.csv")
df
```

```python
df.isnull().sum()
```

```python
df["Region"].mode()
```

```python
df["Region"].fillna('Punjab',inplace = True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df[' Date'].fillna(df[' Date'].mode()[0], inplace = True)
```

```python
df[' Frequency'].mode()
```

```python
df[' Frequency'].fillna('Monthly',inplace = True)
```

```python
df[' Estimated Unemployment Rate (%)'].fillna(df[' Estimated Unemployment Rate (%)'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df[' Estimated Employed'].fillna(df[' Estimated Employed'].mean(), inplace = True)
```

```python
df[' Estimated Labour Participation Rate (%)'].fillna(df[' Estimated Labour Participation Rate (%)'].mean(), inplace = True)
```

```python
df['Area'].mode()
```

```python
df['Area'].fillna('Urban',inplace = True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (14,6))
sns.barplot(x = 'Region', y = ' Estimated Unemployment Rate (%)',data = df)
plt.xticks(rotation = 90)
plt.title('Region - wise Unemployment rate in India')
plt.show()
```

```python
df[' Date'] = pd.to_datetime(df[' Date'])
plt.figure(figsize = (12,6))
plt.plot(df[' Date'],df[' Estimated Unemployment Rate (%)'])
plt.title("Unemployement Rate Over Time")
plt.xlabel(" Date")
plt.ylabel(" Unemployment rate (%)")
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'Area' , y = ' Estimated Unemployment Rate (%)',data = df)
plt.title("Urban vs Rural Unemployement Rate ")
plt.show()
```

```python
numeric_df = df.select_dtypes(include = ['float64','int64'])
plt.figure(figsize = (5,5))
sns.heatmap(numeric_df.corr(),annot = True, cmap = "coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
